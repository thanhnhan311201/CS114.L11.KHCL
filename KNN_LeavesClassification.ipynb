# Tiền xử lý dữ liệu

```
# import một số thư viện cần thiết
import os
import matplotlib.pyplot as plt
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
from collections import deque

```

```

# Đọc file và đưa vào bộ dữ liệu
X_data = np.ones(shape=(1,798768))
y_data = deque()
y_data.append(1)

folder_name = os.listdir("/content/Folio Leaf Dataset/Folio")  # Danh sách tên các class (tên folder sau khi unrar)

for j in range(len(folder_name)):
    path = "/content/Folio Leaf Dataset/Folio" + "/" + folder_name[j]
    # Danh sách tên các tệp hình ảnh trong mỗi class
    file_names = os.listdir(path)
    for i in range(len(file_names)):
        path_f = path + "/" + file_names[i]
        # Đọc tệp hình ảnh,
        image = cv2.imread(path_f)
        # Giảm độ phẩn giải của ảnh xuống để tránh bị crash
        image = cv2.resize(image,(4128//6,2322//6),interpolation = cv2.INTER_AREA)
        # reshape ảnh đang ở dạng ma trận về vector và đưa vào bộ dữ liệu
        image = image.reshape(1, -1)
        X_data = np.vstack((X_data,image))
        y_data.append(j)
```

# Sử dụng model KNN để dự đoán

```
# Chia ngẫu nhiên test data từ data vừa xử lý ở trên
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.33)
```

```
# Khởi tạo tham số cho model
neigh = KNeighborsClassifier(n_neighbors=3)
# fit vào model
neigh.fit(X_train, y_train)
```

```
# Dự đoán
y_predict = neigh.predict(X_test)
```

```
report = classification_report(y_test,y_predict,target_names = folder_name)
```
