```
import json
import numpy as np
from collections import deque

def parse_data(file):
    for l in open(file,'r'):
        yield json.loads(l)

data = list(parse_data('/content/CS114.L11.KHCL/LogisticsRegression/Sarcasm_Project/Sarcasm_Headlines_Dataset.json'))
data = np.array(data)
text = ''
count_words = {}
for i in range(len(data)):
  text += data[i]['headline'] + ' '
text = deque(text)
i = 0
while i < len(text):
  if not((text[i] >='a' and text[i] <='z') or (text[i] >='0' and text[i] <='9') or (text[i] == ' ')):
    text.remove(text[i])
    i-=1
  i+=1
print(text)
```

```
text = text.split()
```

```
for i in range(len(text)):
  count_words[text[i]] = text.count(text[i])
```
